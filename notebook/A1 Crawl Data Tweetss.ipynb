```python
# Crawl Data

filename = 'EfisiensiAnggaran.csv'
# Cuitan pertama kali di 12 Februari dan Pengumuman Inpres No. 1 Tahun 2025 22 Januari 2025
# Surat Sekretaris Mahkamah Agung RI yang terkait dengan efisiensi anggaran pada satuan kerja Mahkamah Agung, yang diterbitkan tanggal 14 Februari 2025
#search_keyword = 'efisiesianggaran since:2025-03-15 until:2025-04-27 lang:id'
search_keyword = '#efisiensianggaran since:2025-02-12 until:2025-05-12 lang:id'
limit = 2000 #batasan scrolling oleh X(elon musk) akun free hanya 600-700/days

!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
df.sample(n=10)
```

```python
# Cek jumlah data yang didapatkan

num_tweets = len(df)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```

```python
#cek data yang missing

import numpy as np
# replace "?" menjadi NaN
df.replace("?", np.nan, inplace = True)
# cek banyaknya missing value pada masing-masing kolom
print(df.isnull().sum())
```
